**Getting the last X games for an Y team **

```python
fbref = sd.FBref(leagues="ENG-Premier League", seasons="2526")

epl_schedule = fbref.read_schedule()
last_5 = epl_schedule[((epl_schedule["home_team"] == "Chelsea") |
                      (epl_schedule["away_team"] == "Chelsea")) & (epl_schedule["score"].notna())]
last_5.tail()
```

```python
ids = last_5["game_id"].tail()
id_list = ids.tolist()
```

```python
id_list
```

**Calling the Defence stats for the last 5 game ids**

```python
all_games_data = []

for match_id in id_list:
    
    print(f"Scraping game: {match_id}")
    
    # Download the data for this specific game
    player_stats = fbref.read_player_match_stats("misc", match_id=match_id)

    cols = [
        ("min", ""),
        ("Performance", "CrdY"),
        ("Performance", "Fls"),
        # ("Performance", "Fld"),
        ("Performance", "TklW"),
    ]

    player_stats_useful = player_stats[cols]

    # flatten the MultiIndex to a single row of names
    player_stats_useful.columns = [
        "min",
        "yellow_cards",
        "fouls_committed",
        # "fouls_against",
        "tackles_won",
    ]

    all_games_data.append(player_stats_useful)
```
